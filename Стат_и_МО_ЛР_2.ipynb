# Задание 1

Представленное распределение являетя гамма-распределением с параметрами $k = 2$, $θ$.

Математическое ожидание: $E(\xi) = \frac{k}{\theta} = \frac{2}{\theta}$
Дисперсия: $D(\xi) =  \frac{k}{\theta^2} = \frac{2}{\theta^2}$
Математическое ожидание квадрата $\xi$: $E(\xi^2) = D(\xi) + (E(\xi))^2 = \frac{6}{\xi^2}$

```python
# Функция построения необходимых графиков

def plot_res(k_values, means, vars, means_sqrt, teor_stats):
  fig, axes = plt.subplots(1, 3, figsize=(18, 5))

  axes[0].plot(k_values, means, 'bo-', markersize=4, linewidth=1)
  axes[0].axhline(y=teor_stats[0], color='r', linestyle='--')
  axes[0].set_xlabel('k')
  axes[0].set_ylabel('Оценка')
  axes[0].set_title('Оценка математического ожидания')
  axes[0].grid(True, alpha=0.3)

  axes[1].plot(k_values, vars, 'bo-', markersize=4, linewidth=1)
  axes[1].axhline(y=teor_stats[1], color='r', linestyle='--')
  axes[1].set_xlabel('k')
  axes[1].set_ylabel('Оценка')
  axes[1].set_title('Оценка дисперсии')
  axes[1].grid(True, alpha=0.3)

  axes[2].plot(k_values, means_sqrt, 'bo-', markersize=4, linewidth=1)
  axes[2].axhline(y=teor_stats[2], color='r', linestyle='--')
  axes[2].set_xlabel('k')
  axes[2].set_ylabel('Оценка')
  axes[2].set_title('Оценка математического ожидания квадрата параметра')
  axes[2].grid(True, alpha=0.3)
```

```python
# Функция построения выборки нужного размера и расчета числовых характеристик. Вызывает функцию plot_res.

def calculus_res(distr, teor_stats):
  means = []
  vars = []
  means_sqrt = []
  k_values = [i for i in range(4, 16)]

  for k in k_values:
    distr_k = distr.rvs(size=2 ** k)

    mean_calc = np.mean(distr_k)
    means.append(mean_calc)
    var_calc = np.var(distr_k)
    vars.append(var_calc)
    mean_sq_calc = np.mean(np.square(distr_k))
    means_sqrt.append(mean_sq_calc)

    print(f"Для выборки размера {2 ** k}:\n  "
         f"E(x): {round(mean_calc, 3)}, "
         f"D(x): {round(var_calc, 3)}, "
         f"E(x^2): {round(mean_sq_calc, 3)}.")

  plot_res(k_values, means, vars, means_sqrt, teor_stats)
```

$θ = 0.5$

$E(\xi) = 4, D(\xi) = 8, E(\xi^2) = 24$

```python
teor_stats = [4, 8, 24]
distr = sps.gamma(a = 2, scale = 1/0.5)
```

```python
calculus_res(distr, teor_stats)
```

$θ = 2$

$E(\xi) = 1, D(\xi) = \frac{1}{2} = 0.5, E(\xi^2) = \frac{3}{2} = 1.5$

```python
teor_stats = [1, 0.5, 1.5]
distr = sps.gamma(a = 2, scale = 1/2)
```

```python
calculus_res(distr, teor_stats)
```

$θ = 8$

$E(\xi) = \frac{1}{4} = 0.25, D(\xi) = \frac{1}{32} = 0.03125, E(\xi^2) = \frac{3}{32} = 0.09375$

```python
teor_stats = [0.25, 0.03125,0.09375]
distr = sps.gamma(a = 2, scale = 1/8)
```

```python
calculus_res(distr, teor_stats)
```

# Задание 2

$$
f_\xi^{λ, a}(x) = \begin{cases}
λe^{-\lambda(x-a)}, x ≥ a, \\
0, else \\
\end{cases}
$$

$(λ, a) = (2, 2)$

Данное распределение представляет собой показательное распределение со сдвигом $a = 2$.

Мода соответствует максимуму функции плотности распределения. На отрезке $x \ge a$ это соответствует $x = a = 2$, т.к. функция $2e^{-2(x-2)}$ является убывающей.

Математическое ожидание можно найти как $\int^{∞}_22xe^{-2(x-2)} = -2\frac{(2x + 1)e^{4-2x}}{4}|^{\infty}_{2} = \frac{5}{2} = 2.5$.

Медиану можно найти как решение уравнения $F(x) = \int^x_22e^{-2(x-2)} = 0.5$.

$\int^x_22e^{-2(x-2)} = -e^{-2(x-2)}|^x_2 = 1 - e^{-2(x-2)} = 0.5$

$e^{-2(x-2)} = 0.5 ⇒ x - 2 = -\frac{ln\frac{1}{2}}{2} = \frac{ln(2)}{2} ⇒ x = 2 + \frac{ln(2)}{2} ≈ 2.3466.$

Таким образом,

$Mo = 2, E = 2.5, Me \approx 2.3466.$

```python
import scipy.stats as sps
import numpy as np
import ipywidgets as widgets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Реализованная самостоятельно функция плотности распределения

class custom_dist(sps.rv_continuous):
    def _pdf(self, x):
        return 2 * np.exp(-2 * (x-2))
```

```python
# Создаем большую и малую выборки на основе custom_dist

distr = custom_dist(a=2, name='custom_distribution')
distr_big = distr.rvs(size=10000)
distr_small = distr.rvs(size=20)
```

```python
# Расчет числовых характеристик

stats_big = [float(sps.mode(distr_big)[0]), float(np.mean(distr_big)), float(np.median(distr_big))]
stats_small = [float(sps.mode(distr_small)[0]), float(np.mean(distr_small)), float(np.median(distr_small))]

stats_teor = [2, 2.5, float(2 + np.log(2)/2)]

print(stats_big)
print(stats_small)
print(stats_teor)
```

```python
x = np.linspace(2, 7, 2000)
y = 2 * np.exp(-2 * (x-2))
```

```python
# Построение графиков

fig, axes = plt.subplots(2, 2, figsize=(18, 7))

axes[0, 0].hist(distr_big, edgecolor = 'black', bins = 100, density = True)
axes[0, 0].axvline(x=stats_big[0], color='r', linestyle='--')
axes[0, 0].axvline(x=stats_big[1], color='y', linestyle='--')
axes[0, 0].axvline(x=stats_big[2], color='g', linestyle='--')
axes[0, 0].set_title('Большая выборка. Оценки моды, М.О., медианы.')

axes[1, 0].hist(distr_small, edgecolor = 'black', bins = 40, density = True, range = (2, 7))
axes[1, 0].axvline(x=stats_small[0], color='r', linestyle='--')
axes[1, 0].axvline(x=stats_small[1], color='y', linestyle='--')
axes[1, 0].axvline(x=stats_small[2], color='g', linestyle='--')
axes[1, 0].set_title('Малая выборка. Оценки моды, М.О., медианы.')

axes[0, 1].hist(distr_big, edgecolor = 'black', bins = 100, density = True)
axes[0, 1].plot(x, y, color = 'cyan', linewidth = 3.0)
axes[0, 1].axvline(x=stats_teor[0], color='r', linestyle='--')
axes[0, 1].axvline(x=stats_teor[1], color='y', linestyle='--')
axes[0, 1].axvline(x=stats_teor[2], color='g', linestyle='--')
axes[0, 1].set_title('Большая выборка. Теоретические значения моды, М.О., медианы.')

axes[1, 1].hist(distr_small, edgecolor = 'black', bins = 40, density = True, range = (2, 7))
axes[1, 1].plot(x, y, color = 'cyan', linewidth = 3.0)
axes[1, 1].axvline(x=stats_teor[0], color='r', linestyle='--')
axes[1, 1].axvline(x=stats_teor[1], color='y', linestyle='--')
axes[1, 1].axvline(x=stats_teor[2], color='g', linestyle='--')
axes[1, 1].set_title('Малая выборка. Теоретические значения моды, М.О., медианы.')


plt.show()
```

# Задание 3

Маргинальные распределения $\xi, \eta$:

$P(\xi = -1) = \frac{2}{5}\sum^{\infty}_{i = 1} \frac{1}{2^i} = \frac{2}{5}, P(\xi = 0) = \frac{1}{5}\sum^{\infty}_{i = 1} \frac{1}{2^i} = \frac{1}{5}, P(\xi = 1) = \frac{2}{5}\sum^{\infty}_{i = 1} \frac{1}{2^i} = \frac{2}{5}$.

$P(\eta = k) = \frac{1}{2^k}$. Это геометричесское распределение при $p = \frac{1}{2}. (P(\eta = k) = \frac{1}{2^k} = (1 - \frac{1}{2})^{k-1}\cdot\frac{1}{2})$.

Найдем ковариационную матрицу:

$M(\xi) = 0, M(\eta) = \frac{1}{p} = 2.$

$M(\xi\eta) = 0.$

$M(\xi^2) = \frac{4}{5}, D(\xi) = \frac{4}{5}$

$D(\eta) = \frac{q}{p} = 2$

$cov(\xi, \eta) = \begin{pmatrix} \frac{4}{5} & 0 \\ 0 & 2 \end{pmatrix}  $.

Найдем кореляционную матрицу:

$r_{\xi, \eta} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} $

```python
import scipy.stats as sps
import numpy as np
import ipywidgets as widgets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 10000000
```

```python
xi_values = np.random.choice([-1, 0, 1], size=N, p=[2/5, 1/5, 2/5]) # Генерируем значения параметра ξ на основе маргинальных вероятностей
eta_values = sps.geom(p=0.5).rvs(size=N) # Генерируем значения параметра η как геометрическое распределение

values = np.column_stack((xi_values, eta_values)) # Создаем матрицу из полученных распределений
```

```python
cor_matrix = np.corrcoef(values.T)
print(cor_matrix)

# Видно, что при достаточно большом n полученная приближенная кореляционная матрица стремиться к теоретической
```
